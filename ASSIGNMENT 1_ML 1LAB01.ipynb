DEFINING FUNCTIONS AND CALLING THEM

```python
#simple calculator using deifferent functions
def calc():
    print("This is a calculator")
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def prod(x,y):
    return x*y
def div(x,y):
    return x//y

calc()
print("The sum of 90 and 45.9027 is: ",add(90,45.9027))
print("The difference of 78 and 10 is: ",sub(78,10))
print("The product of 56 and -178 is: ",prod(56,-178))
print("The quotient when 4 divided by 2 is: ",div(4,2))
```
