**QUESTION 2**

**WRITE A PROGRAM THAT USES 5 DIFFERENT FUNCTIONS AND CALL THEM IN ONE PROGRAM**

```python
#Write a program to build a simple calculator using functions.


def add(x,y):
    c=x+y
    print("The added value is",c)
add(10,15)
def sub(x,y):
    c=x-y
    print("The subtracted value is",c)
sub(50,6)
def mul(x,y):
    c=x*y
    print("The product is",c)
mul(40,6)
def div(x,y):
    c=x/y
    print("The divided value is",c)
div(90,8)
def mod(x,y):
    c=x%y
    print("The modulus is",c)
mod(90,8) 
```
