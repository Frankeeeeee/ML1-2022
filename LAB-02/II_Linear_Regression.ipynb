<b><h3> II. LINEAR REGRESSION</h3></b>

<b><i>Linear Regression:</i></b><br>
<ul>
    <li>Linear regression is a basic and commonly used type of predictive analysis. </li>
<li>The overall idea of regression is to examine two things:
    <ol><li> Does a set of predictor variables do a good job in predicting an outcome (dependent) variable?  </li>
        <li>Which variables in particular are significant predictors of the outcome variable, and in what way do they–indicated by the magnitude and sign of the beta estimates–impact the outcome variable? </li></ol> </li>
<li> These regression estimates are used to explain the relationship between one dependent variable and one or more independent variables. </li>
 <li>The simplest form of the regression equation with one dependent and one independent variable is defined by the formula <i><u>y = m*x + c </u></i>(Simple regression), where y = estimated dependent variable score, c = constant, b = regression coefficient, and x = score on the independent variable.</li></ul>

<b><i>Linear Regression in ML:</i></b><br>
<ul><li>Linear Regression is a machine learning algorithm based on supervised learning. </li>
    <li>It performs a regression task. Regression models a target prediction value based on independent variables.</li>
    <li>It is mostly used for finding out the relationship between variables and forecasting.</li>
</ul>

<b>Scikit-learn</b> is probably the most useful library for machine learning in Python. The sklearn library contains a lot of efficient tools for machine learning and statistical modeling including classification, regression, clustering and dimensionality reduction.

```python
tips
```

```python
tips.columns
#Shows the columns/variables in the dataframe
```

```python
model_df=tips[['total_bill', 'tip']]
#Creating a subset of the dataframe with just total_bill and tip variables.
```

```python
model_df.head()
```

```python
#Setting x and y
x=model_df.total_bill
y=model_df.tip
```

```python
model=LinearRegression()
#assigning just the call to the function to the variable model
```

```python
model.fit(np.array(x).reshape(-1,1),y)
#reshape(-1,1) is used when there is a single feature/array
```

Linear regression fits a straight line or surface that minimizes the discrepancies between predicted and actual output values. There are simple linear regression calculators that use a “least squares” method to discover the best-fit line for a set of paired data.

```python
model.intercept_ 
#This is the value of b0
#The intercept in a regression model represents the mean value of the response variable when all of the predictor variables in the model are equal to zero
```

```python
model.coef_ 
#This is the value of b1
#Gives the coefficient of the features of the dataset
```

```python
model.score(np.array(x).reshape(-1,1),y)
#Score returns the coefficient of determination of the prediction.
#R square value 
#0.4566 signifies a moderately low correlation between bill amount and tips
```

<h4><i>Multiple Linear Regression</i></h4>

When there is one target variable and many x variables, it is a multiple linear regression.<br>
Multiple linear regression (MLR), also known simply as multiple regression, is a statistical technique that uses several explanatory variables to predict the outcome of a response variable. Multiple regression is an extension of linear (OLS) regression that uses just one explanatory variable.

```python
X = df[['acousticness', 'energy', 'danceability']]
y = df['song_popularity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=4)
#train_test_split:
#divides data into training data and testing data

#test_size:
#If float, should be between 0.0 and 1.0 and represent the proportion of the dataset to include in the test split. 
#If int, represents the absolute number of test samples. 
#If None, the value is set to the complement of the train size. 
#If train_size is also None, it will be set to 0.25.

#random_state
#Controls the shuffling applied to the data before applying the split. 
#Pass an int for reproducible output across multiple function calls.
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X_train, y_train)
#This is a low r-square value for the 3 variables and song popularity
```

```python
model.predict(X_test)
#Predicts value of variable passed using the linear model.
```

```python
y_test.values
```

```python
model.score(X_test, model.predict(X_test))
#Correlation between actual values and their predicted values
#Since it is 1.0, the values have a high positive correlation
```

```python
model.score(X_train, model.predict(X_train))
```

```python
model.score(X_test,y_test)
#Correlation between test data
#hardly any correlation
```

```python
model.score(X_train, y_train)
#Correlation between test data
#hardly any correlation
```
