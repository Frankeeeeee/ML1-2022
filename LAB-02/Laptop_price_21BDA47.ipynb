### Multiple Linear Regression Model

```python
LAPPRICE.columns
```

```python
x = LAPPRICE.loc[:,["Inches"]].values

y = LAPPRICE.loc[:,"Price_euros"].values.reshape(-1,1)
```

```python
Sepc1 = LAPPRICE[["Inches"]]
Sepc1
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=4)
```

```python
S = LinearRegression()
S.fit(X_train, y_train)
```

```python
print("beta_0 =",S.intercept_)
print("slopes  = ",S.coef_.round(2),"\n")
```

```python
print("R =",(S.score(X_train, y_train)))
```
