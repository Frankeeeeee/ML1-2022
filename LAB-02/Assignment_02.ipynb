<h2><u>II. Linear Regression </u></h2>

```python
#importing required libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
sns.set_theme(style="darkgrid")
pl17 = sns.regplot(x="total sulfur dioxide",y="density",data=mydata,color="yellow",ci=None)
#There is a slight positive correlation between total sulfur dioxide and the density. 
#Does not imply causation, but shows a trend
```

```python
acidity_data=mydata[['total sulfur dioxide','density','pH','alcohol','quality']]
```

```python
#Setting x and y
x=acidity_data['alcohol']
y=acidity_data.quality
```

```python
m1 = LinearRegression()
m1.fit(np.array(x).reshape(-1,1),y)
```

```python
print("Simple Linear Regression")
print("________________________________________________")
print("Variables: Alcohol and Quality")
print("Equation: y=mx+c")
print("Intercept:",round(m1.intercept_,4))
print("Coefficient:",m1.coef_)
print("R-square:",round(m1.score(np.array(x).reshape(-1,1),y),4))
print("Conclusion: Slight positive correlation between alcohol and quality")
```

```python
m2 = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=4)
```

```python
m2.fit(np.array(x_train).reshape(-1,1),y_train)
```

```python
m2.intercept_
```

```python
m2.coef_
```

```python
m2.score(np.array(x_train).reshape(-1,1),y_train)
```

```python
m2.predict(np.array(x_test).reshape(-1,1))
```
