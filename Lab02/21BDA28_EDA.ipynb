```python
sns.relplot(x = 'Company',y = 'Price_euros',hue = 'Ram',data = df)
plt.xticks(rotation = 45)
```

```python
sns.catplot(x = 'Ram',y = 'Price_euros',data = df)
plt.xticks(rotation = 45)
```

```python
sns.boxplot(x = 'Ram',y = 'Price_euros',data = df)
plt.xticks(rotation = 45)
```

```python
sns.violinplot(x = 'Ram',y = 'Price_euros',data = df,inner = None)
plt.xticks(rotation = 45)
```

```python
sns.barplot(x = 'Ram',y = 'Price_euros',hue = 'TypeName',data = df,palette = "crest")
```

#### From the above graph
#### "2 in 1 Convertible" laptop type with 64GB Ram are the most expensive.
#### Under 8GB and 16GB Ram laptops Workstation types are expensive than the other kinds.
#### 2GB laptops such as the Ultrabook,Notebook,Gaming kinds are the least expensive laptops
