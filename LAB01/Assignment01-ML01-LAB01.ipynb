**Question-2 Calculator**

```python
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
op=input("Enter the operation- +,-,*,/ or %: ")
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    return a//b
def modulus(a,b):
    return a%b
if op=='+':
    addition = add(a,b)
    print ("The sum is: ",addition)
elif op=="-":
        difference=subtract(a,b)
        print("The difference is: ",difference)
elif op=='*':
          product=multiply(a,b)
          print("The product is: ",product)
elif op=='/':
         Quotient=divide(a,b)
         print("The quotient is: ",Quotient)
elif op=='%':
        remainder=modulus(a,b)
        print("The remainder is: ",remainder)
else:
      print("Oops!!Not possible")
```

**Question 3 Read any dataset from https://archive.ics.uci.edu/ml/index.phpand perform a missing value treatment, check for distribution of each variable, find the correlation matrix and write conclusions from the matrix.**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("UsedCarsSA_Clean_EN.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Check for missing values

df.isnull().sum()
```

```python
df = df.rename(columns={"Type": "Model"})
df.head()
```

```python
df = df.drop(['Negotiable'], axis=1)  #Dropping irrelevant columns
df.head()
```

```python
# Summary Statistics for numerical attrubires 
df.describe()
```

```python
df = df[df['Price'] != 0]
df.head()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(df.Price)
plt.ticklabel_format(useOffset=False, style='plain', axis=('x'))

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=df.Price)
plt.ticklabel_format(useOffset=False, style='plain', axis='y')

plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object'])
categorical_columns
```

```python

plt.figure(figsize=(14,8))
plt.subplot(1,3,1)
plt1 = df.Color.value_counts().plot(kind='bar')
plt.title('Colors Histogram')
plt1.set(xlabel = 'Colors', ylabel='Frequency of company')

plt.subplot(1,3,2)
plt1 = df.Fuel_Type.value_counts().plot(kind='bar')
plt.title('Fuel Type Histogram')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of fuel type')

plt.subplot(1,3,3)
plt1 = df.Options.value_counts().plot(kind='bar')
plt.title('Car Options Histogram')
plt1.set(xlabel = 'Car Options', ylabel='Frequency of Car type')

plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(18,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
