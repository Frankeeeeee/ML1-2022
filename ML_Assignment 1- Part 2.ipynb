```python
#W 2. Write 5 different functions and call them in one program.
```

```python
def intro():
    print('What is your name?')
    x = input()
    print('Welcome to the kitchen,',x,'!')
```

```python
def dishname():
    print('\nWhat are we making today?')
    x = input()
```

```python
def dessert():
    print('\nWhat about dessert?')
    x = input()
    print('\nExcellent decision!')
```

```python
def tools():
    print('\nDo you want dry ingredients or wet ingredients?')
    x = input()
    if x == 'dry':
        print('\nLook in the overhead cabinet!')
    elif x == 'wet':
        print('\nLook in the fridge!')
    else:
        print("\nDidn't quite get you. Try again!")
```

```python
def ending():
    print('\nAre you ready to start cooking?')
    x = input()
    if x == 'yes':
        print('\nYay! All the Best!')
    else:
        print("\nYou're a loser")
```

```python
intro()
dishname()
dessert()
tools()
ending()
```
