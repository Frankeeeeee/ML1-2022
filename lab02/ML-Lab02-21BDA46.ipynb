3.log  skweded data

```python
def sigmoid(x):
    e = np.exp(1)
    y = 1 / (1 + e ** (-x))
    return y

d1['fixed acidity'] = np.log(d1['fixed acidity'])

d1['residual sugar'] = sigmoid(d1['residual sugar'])


d1['chlorides'] = np.log(d1['chlorides'])

log_free_sulfur_dioxide = np.log(d1['free sulfur dioxide'].max())
d1['free sulfur dioxide'] = d1['free sulfur dioxide'] ** (1 / log_free_sulfur_dioxide)


log_total_sulfur_dioxide = np.log(d1['total sulfur dioxide'].max())
d1['total sulfur dioxide'] = d1['total sulfur dioxide'] ** (1 / log_total_sulfur_dioxide)


d1['sulphates'] = np.log(d1['sulphates'])

d1.skew()
```
