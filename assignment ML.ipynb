# Can we find correlation between two categorical variables? How?

Yes .We can find correlation between two categorical variables.One statistical test that does this is the Chi Square Test of Independence, which is used to determine if there is an association between two or more categorical variables. if we assume that two variables are independent, then the values of the contingency table for these variables should be distributed uniformly. And then we check how far away from uniform the actual values are.We are not interested in testing the statistical significance however, we are more interested in effect size and specifically in the strength of association between the two variables.
