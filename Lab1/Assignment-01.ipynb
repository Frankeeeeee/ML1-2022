#### 1. Write a program that uses an if-else ladder.

#### 2. Write 5 different functions and call them in one program.

```python
def add(x,y):
    return x+y 
def subtract(x,y):
    return x-y 
def multiply(x,y):
    return x*y 
def power(x,y):
    return x**y 
def divide(x,y):
    return x/y 
def modulus(x,y):
    return x%y

a = int(input("Enter value for a : ")) 
b = int(input("Enter value for b : "))
print("\n1. Add\n2. Subtract\n3. Multiply\n4. Power\n5. Divide\n6. Modulus")

choice = int(input("Enter choice : "))
if choice in [1,2,3,4,5,6]: 
    if choice == 1:
        value = add(a,b)
        print("Added value is : ",value) 
    elif choice == 2:
        value = subtract(a,b) 
        print("Subtracted value is : ",value)
    elif choice == 3:
        value = multiply(a,b) 
        print("Multiplied value is : ",value)
    elif choice == 4: 
        value = power(a,b)
        print("Powered value is : ",value)   
    elif choice == 5:
        value = divide(a,b) 
        print("Divided value is : ",value)
    elif choice == 6:
        value = modulus(a,b) 
        print("Reminder value is : ",value)
    else:
        print("Invalid choice!!")
```

#### 5. Bonus: Can we find correlation between two categorical variables? How?

Checking if two categorical variables are independent can be done with Chi-Squared test of independence.(A chi-square test is a statistical test used to compare observed results with expected results. The purpose of this test is to determine if a difference between observed data and expected data is due to chance, or if it is due to a relationship between the variables you are studying)

This is a typical Chi-Square test: if we assume that two variables are independent, then the values of the contingency table for these variables should be distributed uniformly. And then we check how far away from uniform the actual values are.
