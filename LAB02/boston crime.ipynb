# LINEAR REGRESSION

##### IS THERE A RELATION BETWEEN AGE AND CRIME!

### simple linear regression

```python
y = boston.crim
x = boston.age
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(x).reshape(-1,1),y)
```

```python
model.intercept_ #b0
```

```python
model.coef_ #b1
```

```python
model = sm.OLS(y, x).fit()
```

```python
model.summary()
```
