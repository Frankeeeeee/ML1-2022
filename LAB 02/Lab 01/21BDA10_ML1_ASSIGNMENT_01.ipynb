```python
vowel =  ['a','e','i','o','u']
cons = ['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']
```

```python
str1 = str(input("Enter a letter : "))
if str1 in vowel:
    print("Given letter is a vowel")
elif str1 in cons:
    print("Given letter is consonants")
else:
    print("Entered letter is not a alphabet")
```

```python
def rev(str2):
    res = str2[::-1]
    print(res)
def low(str2):
    res1 = str2.lower()
    print(res1)
def upp(str2):
    res2 = str2.upper()
    print(res2)
def concat(str2 , str3):
    res3 = str2 + str3
    print(res3)

str2 = str(input("Enter a word or sentence : "))
str3 = str(input("Enter a word or sentence : "))

print("\n1. Reverse method\n2. Lowercase method\n3. Upper case method\n4. Concatenation Method")
ch = int(input("Enter a choice : "))

if ch in [1,2,3,4]:
    if ch == 1:
        rev(str2)
    elif ch == 2:
        low(str2)
    elif ch == 3:
        upp(str2)
    elif ch == 4:
        concat(str2 , str3)
else:
    print("Invalid choice")
```
