(B) 2.Write 5 different functions and call them in one program.

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def multiply(a,b):
    return a * b
```

```python
a = int(input("Enter a number: ")) 
b = int(input("Enter another number: ")) 

operator_list = ["add", "sub", "multiply"]
print(operator_list)

operator = input("Enter the operator you want from the operator list: ")

if operator == add:
    print(f"Number1: {a}, Number2: {b}, Operator: add, Answer: {add(a,b)}")
elif operator == sub:
    print(f"Number1: {a}, Number2: {b}, Operator: sub, Answer: {sub(a,b)}")
else:
    print(f"Number1: {a}, Number2: {b}, Operator: multiply, Answer: {multiply(a,b)}") 
```
