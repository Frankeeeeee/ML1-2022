```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Simple Linear Regression

```python
laptop_binary = laptop[['Inches', 'Price_euros']]
```

```python
laptop_binary.columns = ['In', 'Price']
```

```python
laptop_binary.head()
```

```python
sns.lmplot(x ="In", y ="Price", data = laptop_binary, order = 2, ci = None);
```

```python
X = np.array(laptop_binary['In']).reshape(-1, 1)
y = np.array(laptop_binary['Price']).reshape(-1, 1)
```

```python
laptop_binary.dropna(inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
  
plt.show()
```
