# Question1 - Write a program that uses an if-else ladder

```python
mark = int(input("Enter your mark:"))
if mark > 85:
    print("Congrats! You got Distinction")
elif mark>75 and mark<85:
    print("You got First Class")
elif mark>65 and mark<75:
    print("You got Second Class")
elif mark>55 and mark<65:
    print("You got Third Class")
elif mark>45 and mark<55:
    print("Passed the Exam")
else:
    print("You got failed")
```

# Question2 - Write five different functions and call them in one program.

```python
def add(a,b):
    return a+b
def difference(a,b):
    return a-b
def product(a,b):
    return a*b
def quotient(a,b):
    return a/b
def modulud(a,b):
    return a%b
a = int(input("Enter the value of a:"))
b = int(input("Enter the value of b:"))
operator = input("Enter any operator i.e +,-,*,/,%:")
if operator=='+':
    x = add(a,b)
    print("The sum is:",x)
elif operator=='-':
    y = difference(a,b)
    print("The difference is:",y)
elif operator=='*':
    z = product(a,b)
    print("The product is:",z)
elif operator=='/':
    h = quotient(a,b)
    print("The quotient is:",h)
elif operator=='%':
    p = modulus(a,b)
    print("The remainder is:",p)
else:
    print("Invalid Operator")
```
