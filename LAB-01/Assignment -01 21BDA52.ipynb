2. Write five different programs and call them in one program

```python
#Program to create a simple calculator.
def add(a,b):
    return(a+b)
def diff(a,b):
    return(a-b)
def prod(a,b):
    return(a*b)
def div(a,b):
    return (a/b)
def exp(a,b):
    return (a**b)
print("___________________________________________________")
print("                SIMPLE CALCULATOR                  ")
print("___________________________________________________")
x=int(input("Enter the first number: "))
y=int(input("Enter the second number: "))
print("___________________________________________________")
print(" 1. Add \n 2. Subtract \n 3. Multiply \n 4. Divide \n 5. Exponent")
choice=int(input("Enter your choice number: "))
print("___________________________________________________")
if choice==1:
    print("The sum of %s and %s is %s."%(x,y,add(x,y)))
elif choice==2:
    print("The difference of %s and %s is %s."%(x,y,diff(x,y)))
elif choice==3:
    print("The product of %s and %s is %s."%(x,y,prod(x,y)))
elif choice==4:
    print("The quotient after dividing %s by %s is %s."%(x,y,div(x,y)))
elif choice==5:
    print(" %s raised to the power %s is %s."%(x,y,exp(x,y)))
else:
    print("Invalid Input")
```

#### 5. Bonus: Can we find correlation between two categorical variables? How?

Yes, we can find the correlation between two categorical variables using
Chi Square Test.
Method:
Null hypothesis(H0)- two variables are not related to each other
Result of Chi Square Test - probability of H0 being true.

```python
from scipy.stats import chi2_contingency
```

```python
data=pd.read_csv(r"C:\Users\Srinivasa\Downloads\newbla.csv")
data.head(10)
```

```python
CrosstabResult=pd.crosstab(index=data['Location'],columns=data['Bean_Origin'])
print(CrosstabResult)
```

```python
ChiSqResult = chi2_contingency(CrosstabResult)
print(ChiSqResult[1])
```
