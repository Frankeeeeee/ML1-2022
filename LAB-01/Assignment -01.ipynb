<b>5. Bonus: Can we find correlation between two categorical variables? How? </b>

Yes, we can find the correlation between two categorical variables using<br>
Chi Square Test.<br>
<u>Method:</u><br>
Null hypothesis(H0)- two variables are not related to each other<br>
Result of Chi Square Test - probability of H0 being true. 

```python
from scipy.stats import chi2_contingency
```

```python
data2=pd.read_csv(r"C:\Users\Jerusha Gomez\Downloads\adult-stretch.data.csv")
data2.head(10)
```

```python
CrosstabResult=pd.crosstab(index=data2['ADULT'],columns=data2['STRETCH'])
print(CrosstabResult)
```

```python
ChiSqResult = chi2_contingency(CrosstabResult)
print(ChiSqResult[1])
#the value at index 1 is the p-value. 
#The probability of the null hypothesis is high. Meaning there is no relation between the variables.
```
