2.Write 5 different functions and call them in one program.

```python
#calculator
def sum(a,b):
    return a+b

def sub(a,b):
    return a-b

def mult(a,b):
    return a*b

def div(a,b):
    return a/b

def modulo(a,b):
    return a%b

print("Please select operation -\n" \
        "1. Add\n" \
        "2. Subtract\n" \
        "3. Multiply\n" \
        "4. Divide\n"\
        "5. Modulo\n")
  
select = int(input("Select operations form 1, 2, 3, 4, 5 :"))

if select is True:  
    number_1 = int(input("Enter first number: "))
    number_2 = int(input("Enter second number: "))

    if select==1:
        print('sum of two numbers is',(sum(number_1,number_2)))
    elif select==2:
        print('subtraction of two numbers is',(sub(number_1,number_2)))
    elif select==3:
        print('product of two numbers is',(mult(number_1,number_2)))
    elif select==4:
            print('division of two numbers is',(div(number_1,number_2)))
    elif select==5:
            print('modulo of two numbers is',(modulo(number_1,number_2)))
else:
    print("it's invalid input")
```
