```python
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def power(a,b):
    return a**b
def divide(a,b):
    return a/b
```

```python
x = int(input("Enter value for x : "))
y = int(input("Enter value for y : "))

print("\n1. Add\n2. Subtract\n3. Multiply\n4. Power\n5. Divide")
choice = int(input("Enter choice : "))

if choice in [1,2,3,4,5,6]:
    if choice == 1:
        val = add(x,y)
        print("Added value is : ",val)
    elif choice == 2:
        val = subtract(x,y)
        print("Subtracted value is : ",val)
    elif choice == 3:
        val = multiply(x,y)
        print("Multiplied value is : ",val)
    elif choice == 4:
        val = power(x,y)
        print("Powered value is : ",val)
    elif choice == 5:
        val = divide(x,y)
        print("Divided value is : ",val)
    elif choice == 6:
        val = modulus(x,y)
        print("Reminder value is : ",val)
else:
    print("Invalid choice!!")
```
